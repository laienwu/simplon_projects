```python
from sqlalchemy import create_engine
```

```python
# Create an engine (connection to MySQL)
engine = create_engine("mysql+pymysql://root:pswd@localhost:3306/")

# Replace 'db_reviews' with the name of the database you want to create
database_name = 'db_reviews'

# Execute the SQL statement to create the database
with engine.connect() as conn:
    conn.execute(f"CREATE DATABASE IF NOT EXISTS {database_name}")

print(f"Database '{database_name}' created successfully.")
```
